## 휴관일 정보 데이터 만들기 : jsp_holidays

```python
import pandas as pd
holidays = pd.read_csv('./data/jsplib_holidays.csv')
holidays
```

```python
holidays = holidays[2:]
holidays.columns = ['month','holidays_monthly','holiday','holiday_1','holiday_2','holiday_3','holyday_4','holiday_5','holyday_6','holiday_7','holiday_8','holiday_9']
holidays.reset_index(inplace=True)
holidays.drop('index',axis=1,inplace=True)
holidays
```

```python
# jsp_holidays = {key:value for key,value in zip(holidays['month'],holidays['holidays_monthly'])}
# jsp_holidays
```

```python
jsp_holidays = []
for idx,value in enumerate(holidays):
    holiday_values = holidays.loc[idx,'holidays_monthly']
    if holiday_values == 1 : holiday_values = f'0{holiday_values}'
    holiday_month = holidays.loc[idx,'month']
    if len(holiday_month) == 2 : holiday_month = f'0{holiday_month}'
    for holiday in holiday_values.split(" "):
        if len(holiday) == 1 : holiday = f'0{holiday}'
        holiday = f'2023-{holiday_month.replace("월","")}-{holiday}'
        jsp_holidays.append(holiday)
        
jsp_holidays 
```

## 프로그램 집계 범위 만들기
- A_start
- A_end
- B_start
- B_end

```python
import pandas as pd

events_df = pd.read_csv('./output_new/lib_jsplib.csv')
events_df.info()
```

```python
events_df['A_end'] = pd.to_datetime(events_df['start_date'])
events_df['B_start'] = pd.to_datetime(events_df['end_date'])

events_df.info()
```

```python
from datetime import datetime,timedelta
# 1일 프로그램인 경우가 있어서 프로그램 당일을 제외하고, 전날부터 14일, 다음날부터 14일로 잡음
events_df['A_start'] = events_df['A_end'] + timedelta(days=-14)
events_df['A_end'] = events_df['A_end'] + timedelta(days=-1)
events_df['B_start'] = events_df['B_start'] + timedelta(days=1)
events_df['B_end'] = events_df['B_start'] + timedelta(days=13)

events_df.to_csv('./output_new/daterange_check.csv',index=False,encoding='utf-8-sig')
```

```python
print(events_df['A_start'][0])
print(events_df['A_end'][0])
print(events_df['B_start'][0])
print(events_df['B_end'][0])
```

```python
events_df['A_start'] = events_df['A_start'].apply(lambda x:x.strftime("%Y-%m-%d"))
events_df['A_end'] = events_df['A_end'].apply(lambda x:x.strftime("%Y-%m-%d"))
events_df['B_start'] = events_df['B_start'].apply(lambda x:x.strftime("%Y-%m-%d"))
events_df['B_end'] = events_df['B_end'].apply(lambda x:x.strftime("%Y-%m-%d"))
events_df.info()
```

```python
# 휴일 제외하고 더하기
def sum_daily_range(df,filter=jsp_holidays):    
    holidays_in_range = []
    range_sum = 0
    df.reset_index(inplace=True)
    for idx,date in enumerate(df['LON_DE_str']):
        daily_sum = df.loc[idx,'count']
        if date in filter:
            holidays_in_range.append(date)
        else:
            range_sum += daily_sum
    return range_sum, holidays_in_range 
```
