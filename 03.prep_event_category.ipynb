# 대분류로 재분류

- 독서·문예 : 독서 / 퀴즈 / 그림책 / 북클럽 / 북큐레이션 / 저자 / 읽기
- 예술감상·체험 : 전시 / 체험 / 만들기 / 문화
- 정보화·생활교육 : 스마트폰 / 교육 / 교실 / 배우기

```python
temp_df = pd.read_csv('./output/cate_check.csv')
temp_df
```

```python
cate_set = set("//".join(temp_df['cates']).split("//"))
cate_set
```

```python
final_category = ['독서·문예','예술감상·체험','정보화·생활교육']
temp_df['final_category'] = ''
for idx,cate_value in enumerate(temp_df['cates']):
    print(cate_value)
    final_cate = []
    for cate in cate_value:
        if re.search("독서|퀴즈|그림책|북클럽|북큐레이션|저자|읽기",cate_value) : final_cate.append(final_category[0])
        if re.search("전시|체험|만들기|문화",cate_value) : final_cate.append(final_category[1])
        if re.search("스마트폰|교육|교실|배우기",cate_value) : final_cate.append(final_category[2])
        if re.search("수기 확인",cate_value) : final_cate.append("수기 확인")
        final_cate = list(set(final_cate))
    temp_df.loc[idx,'final_category'] = "//".join(final_cate)
    # if idx == 10 : break
temp_df    
```

```python
temp_df['final_category'].value_counts()
```

```python
temp_df.to_csv('./output/cate_check_final03.csv',index=False,encoding='utf-8-sig')
```

```python
for idx,final_cate in enumerate(temp_df['final_category']):
    dummy = temp_df.loc[idx,'events_dummies']
    if final_cate == "독서·문예//예술감상·체험":
        if re.search("읽기",dummy) : temp_df.loc[idx,'final_category'] == final_cate.split("//")[0]
        else : temp_df.loc[idx,'final_category'] == final_cate.split("//")[1]
    if final_cate == "독서·문예//정보화·생활교육//예술감상·체험" : temp_df.loc[idx,'final_category'] = "예술감상·체험"
    if final_cate == "정보화·생활교육//예술감상·체험" : temp_df.loc[idx,'final_category'] = "정보화·생활교육"
    if final_cate == "수기 확인":
        if re.search("편집|특강",dummy): temp_df.loc[idx,'final_category'] = "정보화·생활교육"
        if re.search("만들",dummy): temp_df.loc[idx,'final_category'] = "예술감상·체험"
        else: temp_df.loc[idx,'final_category'] = "정보화·생활교육"
    if re.search("정보화",final_cate): temp_df.loc[idx,'final_category'] = "정보화·생활교육"
    if re.search("예술감상",final_cate): temp_df.loc[idx,'final_category'] = "예술감상·체험"
        
```
